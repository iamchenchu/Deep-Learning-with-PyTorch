#### Random Tensors

Why random tensors ? : these are important because the way neural networks learn is that they start with tensors full of random numbers and then adjust those random numbers to better represent the data

`Start with random numbers -> look at the data -> update random numbers -> look at data -> update random numbers`

Torch random numbers : https://pytorch.org/docs/stable/generated/torch.rand.html

```python
# Create random tensors of shape (6, 7)

random_tensor = torch.rand(6, 7)
print(random_tensor)

print(random_tensor.ndim)
print(random_tensor[0])
```

```python
# Create a random tensor with similar shape to an image tensor
size=(3, 224, 224)

random_image_size_tensor = torch.rand(size) # Height, widht, color channels
print(random_image_size_tensor)
print(random_image_size_tensor.shape)
print(random_image_size_tensor.ndim)
```

#### Tensors of zeros and ones

* The torch default data type is float and it stores it's values in the float32 bits

```python
# Create a tensor of all zeros
zeros = torch.zeros(3, 4)
print(zeros)

print(zeros.ndim)

print(zeros.shape)
```

```python
# Create a tensor of all ones
ones = torch.ones(3, 4)
print(ones)
print(ones.shape)
print(ones.ndim)
```

#### Creating a range of tensors and tensors-like

torch.tensor : https://pytorch.org/docs/stable/tensors.html#torch.Tensor

torch.arange : https://pytorch.org/docs/stable/generated/torch.arange.html



```python
one_to_ten = torch.arange(start=1, end=1000, step=77)
print(one_to_ten)

```

```python
# Creating tensors like  : means we use already created tensors as input to the new tensors as size definition

ten_zeros = torch.zeros_like(input=one_to_ten) # one_to_ten is a size to the ten_zeros tensor
print(ten_zeros)
print(ten_zeros.ndim)
print(ten_zeros.shape)
```

#### Tensor Data Types

Torch.dtype : https://pytorch.org/docs/stable/tensor_attributes.html

**Note :** Tensor data types is one of the 3 big errors you'll run into with Pytorch and deep learning:
  1. Tensors not right datatype
  2. Tensors not right shape
  3. Tensors not on right device

```python
 # Float 32 tensor
float_32_tensor = torch.tensor([3.0, 6.0, 9.0],
                               dtype=None, # What data type is this : float32, float64, float16, etc
                               device=None, # cuda, cpu : what device is your tensor on
                               requires_grad=False # whether or not to track gradients with this tensor operations
                               )

print(float_32_tensor)

print(float_32_tensor.dtype) # returns torch.float32 even though we defined as none because pytorch default is float32

```

```python
float_16_tensor = torch.tensor([12., 13., 17.],
                               dtype=torch.float16,
                               device=None,
                               requires_grad=False)

print(float_16_tensor)
print(float_16_tensor.dtype)
print(float_16_tensor.shape)
print(float_16_tensor.ndim)
```

```python
# We can also conver the 32bit tensor to float16 as per our requirements

float_32_tensor = float_32_tensor.type(torch.float16)
print(float_32_tensor)
print(float_32_tensor.dtype) # now changed to float16
```

```python
float_32_tensor * float_16_tensor  # we can multiply 2 same data type but different size could be multiplied
```

#### Getting information from tensors [Tensor Attributes]

  2. Tensors not right shape - to get the shape of a tensor, we can use `tensor.shape`
  3. Tensors not on right device - to get the device from the tensor, we can use `tensor.device`

#### Manipulating Tensors (Tensors operations)

1. Addition
2. Subtraction
3. Multiplication [Element wise]
4. Division

```python
# Manipulating Tensors (Tensors operations)

# Create tensor and add 10 to it
tensor = torch.tensor([10, 15, 39, 43])
tensor = tensor + 10 # Adds 10 to each element
print(tensor)


# Multiplication by 10

tensor1 = tensor * 10
print(tensor1)

# Subtraction by 100
tensor2 = tensor1 - 100
print(tensor2)

# Try out Pytorch in-built functions
tensor3 = torch.mul(tensor, 10)
tensor3



```

#### Matrix multiplication

Two main ways of performing multiplication in neural networks and deep learning

1. Element-wise multiplication
2. Matrix multiplication (Dot product) : https://www.mathsisfun.com/algebra/matrix-multiplying.html


```python
# Element wise multiplication
tensor = torch.tensor([1, 2, 3])
print(tensor, "*", tensor)
print(f"Equals :{tensor * tensor}")


```

```python
# Matrix Multiplication
print(torch.matmul(tensor, tensor)) # this is
```

```python
%%time
torch.matmul=(tensor, tensor)
```

```python
%%time
value =0
for i in range(len(tensor)):
  value += tensor[i] * tensor[i]

print(value)

# According to this, the inbuilt functions are faster than the loops

```
